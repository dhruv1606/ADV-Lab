```python
# UID: 2021300113
# Name: Dhruv Shah
# ADV Exp-1

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/Online Sales Data.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```python

# Task 1: Basic Plots

# 1. Bar Chart - Product wise sales
plt.figure(figsize=(30, 6))
product_sales = df.groupby('Product Name')['Units Sold'].sum().sort_values(ascending=False)
product_sales.plot(kind='bar')
plt.title('Product Wise Sales')
plt.xlabel('Product Name')
plt.ylabel('Units Sold')
plt.xticks(rotation=90)
plt.show()
```

```python

# 2. Pie Chart - Region wise sales
plt.figure(figsize=(8, 8))
region_sales = df.groupby('Region')['Total Revenue'].sum()
region_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Region Wise Sales')
plt.ylabel('')
plt.show()
```

```python

# 3. Histogram - Distribution of Unit Prices
plt.figure(figsize=(10, 6))
plt.hist(df['Unit Price'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Unit Prices')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()
```

```python

# 4. Timeline Chart - Sales over time
plt.figure(figsize=(10, 6))
df['Date'] = pd.to_datetime(df['Date'])
sales_over_time = df.groupby('Date')['Total Revenue'].sum()
sales_over_time.plot(kind='line', marker='o')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.show()

```

```python
# 5. Scatter Plot - Unit Price vs. Units Sold
plt.figure(figsize=(10, 6))
plt.scatter(df['Unit Price'], df['Units Sold'], color='purple')
plt.title('Unit Price vs Units Sold')
plt.xlabel('Unit Price')
plt.ylabel('Units Sold')
plt.show()
```

```python
# 6. Bubble Plot - Units Sold vs. Total Revenue with Region as color
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Units Sold', y='Total Revenue', size='Units Sold', hue='Region', data=df, alpha=0.6, sizes=(100, 1000))
plt.title('Units Sold vs Total Revenue by Region')
plt.xlabel('Units Sold')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Task 2: Calculate Product-wise sales and Region-wise sales

# Product-wise sales
product_wise_sales = df.groupby('Product Name')['Total Revenue'].sum()
print("Product-wise Sales:")
print(product_wise_sales)
```

```python
# Region-wise sales
region_wise_sales = df.groupby('Region')['Total Revenue'].sum()
print("\nRegion-wise Sales:")
print(region_wise_sales)
```
